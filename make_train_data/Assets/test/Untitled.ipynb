```python
import json
import numpy as np
import torch
import random
from matplotlib import pylab as plt

l1 = 4
l2 = 150
l3 = 100
l4 = 2
model = torch.nn.Sequential(
    torch.nn.Linear(l1,l2),
    torch.nn.ReLU(),
    torch.nn.Linear(l2,l3),
    torch.nn.ReLU(),
    torch.nn.Linear(l3,l4)
)
loss_fn = torch.nn.MSELoss()
learning_rate = 1e-3
optimizer = torch.optim.Adam(model.parameters(),lr = learning_rate)

gamma = 3
epsilon = 1.0



check=1
epoch = 100
losses = []
action_number = 2

for _ in range(epoch):
    with open("situation_car_number.json","r",encoding="utf-8") as file:
        situation = json.load(file)
    situation = np.array(situation)
    situation =situation + np.random.rand(1,4)/10.0
    situation = torch.from_numpy(situation).float()
    check =1
    while(check !=0):
        qval = model(situation)
        qval_ = qval.data.numpy()
        if(random.random() < epsilon):
            action = np.random.randint(0,action_number)
        else:
            action = np.argmax(qval_)
        with open("situation_signal_type.txt","w") as file:
            file.write(str(action))
        
        with open("situation_car_number.json","r",encoding="utf-8") as file:
            situation_2_ = json.load(file)
        situation_2 = np.array(situation_2_)
        situation_2 =situation_2 + np.random.rand(1,4)/10.0
        situation_2 = torch.from_numpy(situation_2).float()

        with open("award.json","r") as file:
            json_file = json.load(file)
            reward = json_file["award"]
            is_crash = json_file["is_crash"]

        with torch.no_grad():
            newQ = model(situation_2)
        maxQ = torch.max(newQ)
        if(is_crash):
            Y = reward
        else:
            Y = reward + (gamma*maxQ)
        Y = torch.Tensor([Y]).detach()
        X = qval.squeeze()[action]
        loss = loss_fn(X,Y)
        optimizer.zero_grad()
        loss.backward()
        losses.append(loss.item())
        optimizer.step()
        situation = situation_2
        check = sum(situation_2_)
    if(epsilon>0.1):
        epsilon -= (1/epsilon)
        
        

    
        
            
```
